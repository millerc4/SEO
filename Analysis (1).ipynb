```python
import numpy as np
import pandas as pd
import math
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data = pd.read_csv('./Data.csv')
print(data.columns)
print(data.head(4))
```

```python
df = pd.DataFrame(data = data)
df["CTR"] = df["CTR"].apply(lambda x: x*100)
```

```python
clicksloc = df.loc[(df["Clicks"] > 0)]
print(clicksloc["Clicks"].mean())
print(clicksloc["Impressions"].mean())
print(clicksloc["CTR"].mean())
print(clicksloc["Position"].mean())
```

```python
impressionsloc = df.loc[(df["Impressions"] > 0)]
print(impressionsloc["Clicks"].mean())
print(impressionsloc["Impressions"].mean())
print(impressionsloc["CTR"].mean())
print(impressionsloc["Position"].mean())
```

```python
clickshead = clicksloc.head(10)
sns.set(style="whitegrid")
ax = sns.barplot(x="Query", y="Clicks", data=clickshead)
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.show()
```

```python
bx = sns.barplot(x="Query", y="Impressions", data=clickshead)
bx.set_xticklabels(bx.get_xticklabels(), rotation=40, ha="right")
plt.show()
```

```python
cx = sns.barplot(x="Query", y="CTR", data=clickshead)
cx.set_xticklabels(cx.get_xticklabels(), rotation=40, ha="right")
plt.show()
```

```python
dx = sns.barplot(x="Query", y="Position", data=clickshead)
dx.set_xticklabels(dx.get_xticklabels(), rotation=40, ha="right")
plt.show()
```

```python
#Graph for impressions greater than 1000 but 0 clicks
optloc = df.loc[(df["Impressions"] > 1000) & (df["Clicks"] == 0)]
opthead = optloc.head(25)
ix = sns.barplot(x="Query", y="Impressions", data=opthead)
ix.set_xticklabels(ix.get_xticklabels(), rotation=50, ha="right")
plt.show()
```

```python
#Graph for clicks but low impressions
optloc2 = df.loc[(df["Clicks"] > 0) & (df["Impressions"] < 100)]
jx = sns.barplot(x="Query", y="Clicks", data=optloc2)
jx.set_xticklabels(jx.get_xticklabels(), rotation=50, ha="right")
plt.show()
```

```python
#Graph for high clicks and impressions but bad position
posloc = df.loc[(df["Clicks"] > 0) & (df["Impressions"] > 500)]
postail = posloc.sort_values(by="Position")
kx = sns.barplot(x="Query", y="Position", data=postail)
kx.set_xticklabels(kx.get_xticklabels(), rotation=50, ha="right")
plt.show()
```
